4) Missing Values

```python
df['marks']=df['marks'].fillna(df['marks'].mean())
```

5. Find Top & Bottom Performers

```python
Avg_marks=df.groupby('name')['marks'].mean().sort_values(ascending=False)
print(Avg_marks)
print(Avg_marks.idxmax())
print(Avg_marks.idxmin())   

```
